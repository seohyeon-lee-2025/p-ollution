```python
import pandas as pd
import numpy as np
import os
import re
```

```python
directory = "data"

#loop through data folder and open each file as csv. take name from disease code
df_list = []
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)

    # checking if it is a file
    if os.path.isfile(f):
        pattern = r'([A-Za-z]\d{3})'
        match = re.search(pattern, f)
        title = match.group(1) 
        title = pd.read_csv(f)
        df_list.append(title)
        #print(title[:4])
```

```python
#concat all disease datasets below each other
long_df = pd.concat(df_list, axis = 0)
```

```python
long_df.columns
long_df.head()
```

```python
#long_df.head()
long_df.rename(columns = {"Unnamed: 0": "disease_code", "yr-month": "region"}, inplace = True)
```

```python
#save this column to concat later
disease_code = long_df["disease_code"]
region = long_df["region"]
```

```python
region_lst = region.tolist()
```

```python
#filter columns in format "2017년 1월". these contain the number of patients for each disease
pattern = r'\d{4}년 \d{2}월'

# Use the filter method to keep only columns matching the pattern
filtered_df = long_df.filter(regex=pattern, axis=1)
filtered_df.head()
```

```python
column_values = pd.Series(region_lst)
filtered_df.insert(loc=0, column='region', value=column_values)
filtered_df.head()
```

```python
filtered_df.set_index(disease_code, inplace=True, drop = True)
filtered_df.head()
```

```python
filtered_df.drop(['disease_code'], axis = 0, inplace=True)
```

```python
filtered_df.rename(columns=lambda x: x.replace('년 ', '-').replace('월', ''), inplace=True)
filtered_df.columns
```
