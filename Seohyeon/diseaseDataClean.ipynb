```python
import pandas as pd
import numpy as np
import os
import re
```

```python
directory = "disease_data"

#loop through data folder and open each file as csv. take name from disease code
df_list = []
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)

    # checking if it is a file
    if os.path.isfile(f):
        pattern = r'([A-Z]\d{3})'
        match = re.search(pattern, f)
        title = match.group(1) 
        title = pd.read_excel(f)
        df_list.append(title)
        #print(title[:4])
```

```python
#concat all disease datasets below each other
long_df = pd.concat(df_list, axis = 0)
```
