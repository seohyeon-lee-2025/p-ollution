```python
import pandas as pd
import numpy as np
import os
import re
```

```python
directory = "air_data"

#loop through data folder and open each file as csv. take name from disease code
df_list = []
title_list = []
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)

    # checking if it is a file
    if os.path.isfile(f):
        title = filename.split('.')[0]
        title_list.append(title)
        if filename.split('.')[1] == "csv":
            df = pd.read_csv(f)
           
        elif filename.split('.')[1] == "xlsx":
            df = pd.read_excel(f)
        df_list.append(df)
```

```python
df_dict = list(zip(df_list, title_list))

for df, title in df_dict:
    cols = df.columns
    new_cols = cols.map(lambda x: x + " " + title)
    df.columns = new_cols
    print(df.head())
```
