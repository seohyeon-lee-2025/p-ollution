```python
import pandas as pd
import os
```

```python
#open everything in precocid
directory = "rawdata/air_data_precovid"

#loop through data folder and open each file as csv. take name from disease code
df_list = []
title_list = []
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)

    # checking if it is a file
    if os.path.isfile(f):
        print(filename)
        title = filename
        print(title)
        title_list.append(title)
        if filename.split('.')[1] == "csv":
            df = pd.read_csv(f) 
            #print(df.head())  
        elif filename.split('.')[1] == "xlsx":
            df = pd.read_excel(f)
            #print(df.head())  
        #append df to df list regardless of f type
        df.rename( columns={'Unnamed: 0' :'year_month'}, inplace=True)
        year_month = df['year_month']
       # set first column as index
        df.reset_index(year_month, drop = True, inplace = True)
        print(df.head())
        df_list.append(df)
        #print(title[:4])
```
