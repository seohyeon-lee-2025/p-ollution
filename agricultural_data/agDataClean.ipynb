```python
import pandas as pd
import os
import re
```

```python
#script to clean everything in agricultural data

directory = "agricultural_data"

#loop through data folder and open each file as csv. take name from disease code
df_list = []
title_list = []
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)

    # checking if it is a file
    if os.path.isfile(f):
        df = pd.read_excel(f)
        if filename != "main_crop_production.xlsx":
            df_list.append(df)
            title_list.append(filename)

```

```python
for title, df in zip(title_list, df_list):
    if title != 'main_crop_production':
        #takeproduction only from each year (filter column names that end with ".1")
        pattern = r".*\.1$"
        filtered_cols = [col for col in df.columns if re.match(pattern, col)]
        filtered_df = df[filtered_cols]
        transposed_df = filtered_df.transpose()
```

```python
# Concatenate the datasets along rows
combined_df = pd.concat(df_list, ignore_index=True)

# Group by the columns and sum the values
result_df = combined_df.groupby(combined_df.columns, axis=1).sum()

# If you have NaN values due to missing data in some datasets, you can replace them with 0
result_df = result_df.fillna(0)

print(result_df)
```
